```python
import pandas as pd
```

```python
# Replace 'your_file.csv' with the actual name of your uploaded file
df = pd.read_csv('/content/denmark v norway p parties - Sheet2.csv')
```

```python
df
```

```python
leaning_counts = df.groupby(['Country', 'Leaning'])['Leaning'].count().reset_index(name='Count')
print(leaning_counts)
```

```python
leaning_counts_df = pd.DataFrame(leaning_counts)
```

```python
filtered_df = leaning_counts_df[leaning_counts_df['Leaning'] != 'Single issue']
```

```python
filtered_df
```

```python
filtered_df = leaning_counts_df[~leaning_counts_df['Leaning'].str.contains('Single issue', case=False, na=False)]
```

```python
filtered_df
```

```python
import matplotlib.pyplot as plt

# Filter data for Denmark and Norway
denmark_data = filtered_df[filtered_df['Country'] == 'Denmark']
norway_data = filtered_df[filtered_df['Country'] == 'Norway']

```

```python
# Filter data for Norway using str.contains to handle potential whitespace
norway_data = filtered_df[filtered_df['Country'].str.contains('Norway', case=False, na=False)]
```

```python
# Create pie charts
fig, axes = plt.subplots(figsize=(12, 6))  # 1 row, 2 columns for subplots

# Norway pie chart
axes.pie(norway_data['Count'], labels=norway_data['Leaning'], autopct='%1.1f%%', startangle=90)
axes.set_title('Political Leanings in Norway')

plt.show()  # Display the charts
```

```python
# Create pie charts
fig, axes = plt.subplots(figsize=(12, 6))  # 1 row, 2 columns for subplots

# Denmark pie chart
axes.pie(denmark_data['Count'], labels=denmark_data['Leaning'], autopct='%1.1f%%', startangle=90) # Changed labels to denmark_data['Leaning']
axes.set_title('Political Leanings in Denmark')

plt.show()  # Display the charts
```
